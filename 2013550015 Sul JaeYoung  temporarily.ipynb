```python
import  pybinding  as pb
import numpy as np
import matplotlib.pyplot as plt
from pybinding.repository import graphene
from matplotlib.collections import Collection
from matplotlib.artist import allow_rasterization
from math import pi, sqrt
```

```python
a_cc = graphene.a_cc
a = graphene.a
```

```python
def vacancy(position, radius):
    @pb.site_state_modifier
    def modifier(state, x, y):
        x0, y0 = position
        state[(x-x0)**2 + (y-y0)**2 < radius**2] = False
        return state
    return modifier
```

```python
model = pb.Model(
    graphene.monolayer_4atom(),
    pb.rectangle(x=3, y=0.3), 
    #vacancy(position=[0.5, 0], radius=0.1),
    #vacancy(position=[0, 0], radius=0.1),
    pb.translational_symmetry(a1=1, a2=False)
)
plt.figure(figsize=[10,20])
model.plot()
model.lattice.plot_vectors(position=[0.6, -0.25])
plt.show()

plt.figure()
plt.title('Eigenvalues')
solver = pb.solver.lapack(model)
a = graphene.a_cc * sqrt(3)  # ribbon unit cell length
bands = solver.calc_bands(-pi/a, pi/a)
bands.plot()
plt.show

plt.figure()
plt.title('Density of state')
dos = solver.calc_dos(energies=np.linspace(-7,7, 200), broadening=0.05)  # [eV]
dos.plot()
plt.show()
```
